```python
client = cowtools.jobqueue.GetCondorClient(max_workers=150)
client
```

```python
with client:
    MC_ddc.do_preprocess(output_file="xcache_test_fileset",
                                   step_size=10000,
                                   align_to_clusters=True,
                                   scheduler_url=client.scheduler_info()["address"])
```
